```python
msft_y=web.get_data_yahoo("MSFT",start="2019-06-06",end="2020-06-06")
msft1=web.get_data_yahoo("MSFT",start="2020-05-06",end="2020-06-06")
```

```python
fig,ax=plt.subplots(figsize=(15,5))
ax.plot(msft_y.Close)
ax.set_xlabel("Date")
ax.set_ylabel("Close")
ax.plot(msft1['Close'])

ax2=ax.twinx()
ax2.set_ylabel("Last 1 month",fontsize=14)
plt.show()
# save the plot as a file
fig.savefig('msft.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
